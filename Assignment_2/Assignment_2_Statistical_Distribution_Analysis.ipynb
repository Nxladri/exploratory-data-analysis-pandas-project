```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart_diseases.csv');
```

Statistical plots & distribution analysis


```python
#Draw histograms, KDEs and boxplots to inspect & distributions.
df.hist(figsize=(20,20))
heart_disease_histograms = sns.histplot(data=df, bins=20, kde=True, color='skyblue')
plt.show()
plt.savefig('heart_disease_histograms.png')
```

```python
# draw the kernel density estimation (KDE) plot
plt.figure(figsize=(10, 6))
plt.title('Kernel Density Estimation (KDE) Plot')
plt.xlabel('Value Range')
plt.ylabel('Density')
heart_disease_kde = sns.kdeplot(data=df, fill=True, color='olive')
plt.show()
plt.savefig('heart_disease_kde.png')
```

```python
# display the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, color='lightcoral')
plt.title('Boxplot of Heart Disease Data')
plt.xlabel('Features')
plt.ylabel('Values')
heart_disease_boxplot = sns.boxplot(data=df, color='lightcoral')
plt.show()
plt.savefig('heart_disease_boxplot.png')
```

```python
#Compare distributions across groups (e.g., region A vs B).
# Distribution of cholesterol levels across age
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='age', y='chol')
plt.title('Cholesterol level by age')
plt.xlabel('Age')
plt.ylabel('chol')
heart_disease_group_comparison1 = sns.histplot(data=df, x='age', y='chol')
plt.show()
plt.savefig('heart_disease_group_comparison1.png')
```

```python
# Distribution of chest pain across age
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='age', y='cp')
plt.title('Chest Pain across age')
plt.ylabel('Chest Pain')
plt.xlabel('Age')
heart_disease_group_comparison2 = sns.histplot(data=df, x='age', y='cp')
plt.show()
plt.savefig('heart_disease_group_comparison2.png')
```

```python
# Distribution of thalachemia across age
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='age', y='thalach')
plt.title('thalachemia across age')
plt.ylabel('thalachemia')
plt.xlabel('Age')
heart_disease_group_comparison3 = sns.histplot(data=df, x='age', y='thalach')
plt.show()
plt.savefig('heart_disease_group_comparison3.png')
```

```python
# Detect outliers and comment on skewness /spread.
# outlier detection
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, color='lightcoral')
sns.despine()
plt.title('Outliers spreadness of the dataset')
plt.xlabel('Features')
plt.ylabel('Values')
heart_disease_outliers = sns.boxplot(data=df, color='lightcoral')
plt.show()
plt.savefig('heart_disease_outliers.png')
```

```python
# Evaluate the skewness of the dataset
skewness = df.skew()
print("Skewness of the dataset:")
print(skewness)
```
