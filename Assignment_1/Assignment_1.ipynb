```python
import pandas as pd
import numpy as np
```

```python
#Read CSV/Excel into DataFrame and inspect head/tail/types.
df = pd.read_csv('heart_diseases.csv')
print('This is the head of the DataFrame:\n')
print(df.head())
```

```python
print('This is the tail of the DataFrame:\n')
print(df.tail())
```

```python
print('These are the data types of the DataFrame:\n')
print(df.dtypes)
```

```python
#Compute summary stats (mean, median,min, max, count)....only these and exclude 25% and 75% percentiles.
print('These are the summary statistics of the DataFrame:\n')
print('Mean:\n', np.mean(df))
print('Median:\n', np.median(df))
print('Min:\n', np.min(df))
print('Max:\n', np.max(df))
print('Count:\n', df.count())
```

```python
#Filter rows, select columns and slice,subsets.
print('Filtered rows where age > 50:\n')
f_rows = df[df['age'] > 50]
print(f_rows)
```

```python
# #Filter rows, select columns and slice,subsets.
print('Filter rows where sex is 1 and chest pain type is 3:\n')
f_rows1 = df[(df['sex'] == 1) & (df['cp'] == 3)]
print(f_rows1)
```

```python
# #Filter rows, select columns and slice,subsets.
print('Select columns age and chol:\n')
selected_cols = df[['age', 'chol']]
print(selected_cols)
```

```python
print('Slice index 10 from 20: \n')
slice = df.iloc[10:21]
print(slice)
```

```python
# subsets of the DataFrame based on conditions
print('Subset of rows where age > 50 and sex is 0:\n')
sub = df[(df['age'] > 50 ) & (df['sex'] == 0)]
print(sub)
```

```python
# save the filtered result to a new CSV file
sub.to_csv('filtered_data.csv', index=False)
```
